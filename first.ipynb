### б) объёмов первых 5 уровней стaканов

Функция для получения колонок с данными про уровни стакана

```python
def five_level_columns(name, lobs_type):
    columns = []
    for i in range(5):
        column = str(name) + ":Binance:LinearPerpetual_" + str(lobs_type) + "_vol_" + str(i)
        columns.append(column)
    return columns
```

Кумулятивные суммы по уровням

```python
levels_ask_btc = lobs_btc[five_level_columns('btcusdt', 'ask')].cumsum(axis=1)
levels_ask_eth = lobs_eth[five_level_columns('ethusdt', 'ask')].cumsum(axis=1)
levels_bid_btc = lobs_btc[five_level_columns('btcusdt', 'bid')].cumsum(axis=1)
levels_bid_eth = lobs_eth[five_level_columns('ethusdt', 'bid')].cumsum(axis=1)
```

```python
for i, column in enumerate(five_level_columns('btcusdt', 'ask')):
    plt.hist(np.log(levels_ask_btc[column]))
    plt.xlabel('log(Объем)')
    plt.title('Объем первых ' + str(i) + ' уровней стакана для btcusdt, ask')
    plt.show()
```

```python
for i, column in enumerate(five_level_columns('btcusdt', 'bid')):
    plt.hist(np.log(levels_bid_btc[column]))
    plt.xlabel('log(Объем)')
    plt.title('Объем первых ' + str(i) + ' уровней стакана для btcusdt, bid')
    plt.show()
```

```python
for i, column in enumerate(five_level_columns('ethusdt', 'ask')):
    plt.hist(np.log(levels_ask_eth[column]))
    plt.xlabel('log(Объем)')
    plt.title('Объем первых ' + str(i) + ' уровней стакана для ethusdt, ask')
    plt.show()
```

```python
for i, column in enumerate(five_level_columns('ethusdt', 'bid')):
    plt.hist((levels_bid_eth[column]))
    plt.xlabel('log(Объем)')
    plt.title('Объем первых ' + str(i) + ' уровней стакана для ethusdt, bid')
    plt.show()
```

```python
for i, column in enumerate(five_level_columns('ethusdt', 'bid')):
    plt.hist(np.log(levels_bid_eth[column]))
    plt.xlabel('log(Объем)')
    plt.title('Объем первых ' + str(i) + ' уровней стакана для ethusdt, bid')
    plt.show()
```

### График величины "пробития" стакана от размера трейда того, насколько "пробивается" стакан в результате трейда

Соберем данные вместе. Вначале получим нужные колонки из таблиц

```python
trades_eth_bid = trades_eth[trades_eth.aggro_side == 'BID'][['exchange_ts', 'size']]
trades_eth_ask = trades_eth[trades_eth.aggro_side != 'BID'][['exchange_ts', 'size']]
```

```python
trades_btc_bid = trades_btc[trades_btc.aggro_side == 'BID'][['exchange_ts', 'size']]
trades_btc_ask = trades_btc[trades_btc.aggro_side != 'BID'][['exchange_ts', 'size']]
```

```python
lobs_eth_bid = lobs_eth[[' exchange_ts', 'ethusdt:Binance:LinearPerpetual_bid_price_0']]
lobs_eth_ask = lobs_eth[[' exchange_ts', 'ethusdt:Binance:LinearPerpetual_ask_price_0']]
```

```python
lobs_btc_bid = lobs_btc[[' exchange_ts', 'btcusdt:Binance:LinearPerpetual_bid_price_0']]
lobs_btc_ask = lobs_btc[[' exchange_ts', 'btcusdt:Binance:LinearPerpetual_ask_price_0']]
```

Отсортируем их по дате, чтобы можно было смэтчить

```python
trades_eth_bid = trades_eth_bid.sort_values(by=['exchange_ts']).set_index("exchange_ts")
trades_eth_ask = trades_eth_ask.sort_values(by=['exchange_ts']).set_index("exchange_ts")
trades_btc_bid = trades_btc_bid.sort_values(by=['exchange_ts']).set_index("exchange_ts")
trades_btc_ask = trades_btc_ask.sort_values(by=['exchange_ts']).set_index("exchange_ts")

lobs_eth_bid = lobs_eth_bid.sort_values(by=[' exchange_ts']).set_index(" exchange_ts")
lobs_eth_ask = lobs_eth_ask.sort_values(by=[' exchange_ts']).set_index(" exchange_ts")
lobs_btc_bid = lobs_btc_bid.sort_values(by=[' exchange_ts']).set_index(" exchange_ts")
lobs_btc_ask = lobs_btc_ask.sort_values(by=[' exchange_ts']).set_index(" exchange_ts")
```

```python
eth_bid = pd.concat([trades_eth_bid, lobs_eth_bid], sort=False)
eth_bid
```

```python
len(eth_bid) == len(trades_eth_bid) + len(lobs_eth_bid)
```

график величины "пробития" стакана от размера трейда того, насколько "пробивается" стакан в результате трейда для eth, bid

```python
prev_price, flag = 0, 0
prices, sizes = [], []

for i in range(len(eth_bid)):
    cur_str = eth_bid.iloc[i]
    if np.isnan(cur_str['size']): #trade
        size = cur_str['size']
        flag = 1
    else:
        if flag:
            sizes.append(size)
            prices.append(cur_str['ethusdt:Binance:LinearPerpetual_bid_price_0'] - prev_price)
            flag = 0
        prev_price = cur_str['ethusdt:Binance:LinearPerpetual_bid_price_0']
```
